```python
import pandas as pd
```

```python
df=pd.read_csv("ESTATEFP.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
x=df.drop("Name",axis=1)
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
v=VarianceThreshold(0)
```

```python
x=v.fit_transform(x)
```

```python
x=pd.DataFrame(x,columns=v.get_feature_names_out())
```

```python
x["label"]=df["Name"]
```

```python
x.shape
```

```python
x.to_csv("ESTATFP_BEST.csv",index=None)
```
