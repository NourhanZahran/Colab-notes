```python
from pubchempy import PubChemHTTPError
from tqdm import tqdm
import pubchempy as pcp
```

```python
smiles = []

for i in tqdm(pc["cid"]):
    retry_attempts = 3  # You can adjust the number of retry attempts as needed

    while retry_attempts > 0:
        try:
            mol = pcp.Compound.from_cid(int(i))
            smi = mol.canonical_smiles
            smiles.append(smi)
            break  # Break out of the loop if successful
        except PubChemHTTPError as e:
            if 'PUGREST.ServerBusy' in str(e):
                print(f"Server busy for CID {i}. Retrying...")
                retry_attempts -= 1
            else:
                print(f"Error for CID {i}: {e}")
                break  # Break if the error is not 'ServerBusy'
        except Exception as e:
            print(f"Error for CID {i}: {e}")
            break  # Break for other exceptions

    if retry_attempts == 0:
        print(f"Failed to retrieve CID {i} after multiple attempts.")
```

```python
smiles = []

for i in tqdm(pc["cid"]):

    mol = pcp.Compound.from_cid(int(i))
    smi = mol.canonical_smiles
    smiles.append(smi)
```
