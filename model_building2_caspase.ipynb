```python
import pandas as pd
```

```python
df = pd.read_csv("PubChemFP_best.csv")


df.head()
```

```python
X = df.drop("label",axis=1)

y = df["label"]

```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
from sklearn.model_selection import cross_val_score

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score,recall_score,precision_score,matthews_corrcoef,roc_auc_score,make_scorer,auc
```

```python
rf = RandomForestClassifier(class_weight="balanced",random_state=12)
```

```python
accuracy = cross_val_score(rf, X_train, y_train, scoring='accuracy', cv = 10)
print(accuracy)
```

```python
print("Accuracy of rf with Cross Validation is:",accuracy.mean() * 100)
```

```python
precision = cross_val_score(rf, X_train, y_train, scoring='precision', cv = 10)
print(precision)
```

```python
print("Precision of rf with Cross Validation is:",precision.mean() * 100)
```

```python
recall = cross_val_score(rf, X_train, y_train, scoring='recall', cv = 10)
print(recall)
```

```python
print("recall of rf with Cross Validation is:",recall.mean() * 100)
```

```python
rf.fit(X_train,y_train)
```

```python
yHat_train= rf.predict(X_train)

yHat_test = rf.predict(X_test)

```

```python
print(recall_score(y_train,yHat_train)) #SN
print(recall_score(y_train,yHat_train,pos_label=0)) #S
print(accuracy_score(y_train,yHat_train)) # Q
```

```python
print(recall_score(y_test,yHat_test))
print(recall_score(y_test,yHat_test,pos_label=0))
print(accuracy_score(y_test,yHat_test))
```

```python
RocCurveDisplay.from_estimator(rf,X_test,y_test)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("Sub_BEST.csv")


X = df.drop("label",axis=1)

y = df["label"]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
from sklearn.model_selection import cross_validate

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score,recall_score,precision_score,matthews_corrcoef,roc_auc_score,make_scorer,auc
```

```python
SN = make_scorer(recall_score)

SP = make_scorer(recall_score,pos_label=0)


Q = make_scorer(accuracy_score)

MCC = make_scorer(matthews_corrcoef)


scoring = {"SN":SN,"SP":SP,"Q":Q,"MCC":MCC}
```

```python
rf = RandomForestClassifier(class_weight="balanced",random_state=12)
```

```python
cv_results = pd.DataFrame(cross_validate(rf,X_train,y_train,cv=10,scoring=scoring))
```

```python
cv_results.describe().T
```

```python
rf.fit(X_train,y_train)
```

```python
yHat_train= rf.predict(X_train)

yHat_test = rf.predict(X_test)

```

```python
print(recall_score(y_train,yHat_train)) #SN
print(recall_score(y_train,yHat_train,pos_label=0)) #SP
print(matthews_corrcoef(y_train,yHat_train)) #MCC
print(accuracy_score(y_train,yHat_train)) # Q
```

```python
print(recall_score(y_test,yHat_test))
print(recall_score(y_test,yHat_test,pos_label=0))
print(matthews_corrcoef(y_test,yHat_test))
print(accuracy_score(y_test,yHat_test))
```

```python
from sklearn.metrics import RocCurveDisplay
```

```python
RocCurveDisplay.from_estimator(rf,X_test,y_test)
```
