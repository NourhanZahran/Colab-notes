```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
```

```python
# using the RDKit cheminformatics library in Python to read molecule information from "HSD1_binary.smi" and create a molecule supplier object (t1)
```

```python
t1=Chem.SmilesMolSupplier("HSD1_binary.smi", delimiter="\t", titleLine=False)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, callbacks
from sklearn.preprocessing import LabelEncoder
```

```python
#creating two empty lists, FP and IDs, to store molecular fingerprints and molecule IDs
# Generates a Morgan fingerprint for the molecule (mol) using RDKit's GetMorganFingerprintAsBitVect function
#FP list will contain Morgan fingerprints for all valid molecules, and the IDs list will contain the corresponding molecule IDs
```

```python
FP, IDs = [], []
for i, mol in enumerate(t1):
    if mol is not None:
        FP.append(AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048))
        IDs.append(mol.GetProp("_Name"))
```

```python
#converting a list of RDKit Morgan fingerprints (FP) and  molecule IDs (IDs) into NumPy arrays
# ToBitString This method returns a binary string representing the presence or absence of features in the fingerprint
```

```python
X = np.array([list(fp.ToBitString()) for fp in FP], dtype=np.float32)
y = np.array(IDs)
```

```python
#LabelEncoder to transform a list of categorical labels into numerical representations.
#The transformed labels are stored in the variable y_encoded.
#y_encoded will be a NumPy array containing the numerical representations of the original categorical labels

```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#preparation of data for a deep learning model, involving splitting the dataset into training and testing sets
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=12)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#The reshaping is done to transform each fingerprint into a 2D array with dimensions (2048, 1)
#X_train.shape[0] The first dimension of X_train, which represents the number of samples (rows).
```

```python
# Reshape the input data to match CNN input requirements
X_train = X_train.reshape(X_train.shape[0], 2048, 1)
X_test = X_test.reshape(X_test.shape[0], 2048,1)
```

```python
#defines a convolutional neural network (CNN) model using the Keras library
#Dropout helps prevent overfitting
```

```python
model = keras.Sequential([
    layers.Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(2048, 1)),
    layers.MaxPooling1D(pool_size=2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(len(label_encoder.classes_), activation='sigmoid')  # Number of units matches the number of classes
])
```

```python
#The compile method in Keras is used to configure the learning process of a neural network model
#Optimizer; algorithm commonly used for training neural networks
#binary_crossentropy': This loss function is commonly used for binary classification problems. It measures the difference between the true values and the predicted values for each sample
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Training will be stopped when the validation loss stops improving.
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

```

```python
model.summary()
```

```python
# Train the model with early stopping
model.fit(X_train, keras.utils.to_categorical(y_train), epochs=100, batch_size=32, validation_data=(X_test, keras.utils.to_categorical(y_test)), callbacks=[early_stopping])

```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test, keras.utils.to_categorical(y_test))
print("Test Accuracy:", accuracy)
```
